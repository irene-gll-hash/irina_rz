```python
matrix = np.array([1,3,5,7,6,5,6,8,7])
matrix = matrix.reshape(3,3)
print(matrix)
print('shape:', matrix.shape)
print('dtype:', matrix.dtype)
```

```python
# выводим второй и последний элемент списка
print(n_list[1])
print(n_list[-1])
```

```python
# выводим со второго до пятого невключительно элементы списка
print(n_list[2:5])
```

```python
# срезом получаем второй элемент второй строки
print(matrix[1,1])
```

```python
# срезом получаем вторую строку
print(matrix[1:2])
```

```python
# выводим все строки первого столбца
print(matrix[:, 0])
```

```python
# выводим все столбцы первой строки
print(matrix[0, :])
```

```python
# выводим подмассив
print(matrix[0:2, 0:2])
```
